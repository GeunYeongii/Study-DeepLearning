## 1. python 문법

3주차에 배웠던 np.arange와 비슷한 패턴이 반복됩니다.

처음 배울 때가 어렵지 초반에 일부 개념들을 확실히 익혀두면 점점 쉬워집니다.

```python
a = np.arange(36).reshape(6,6)
a
```

### 1) Indexing

**<font color="red"> [start] => start만** 

```python
a[3]
```

```python
a[3][5]
```

### 2) slicing

#### 1> 기본 

**<font color="red"> [start:end] => start 이상 end 미만** 

```python
a[1:5]
```

#### 2> step

**<font color="red"> [start:end:step] => start 이상 end 미만(step 만큼 증가)** 

```python
a[1:5:2]
```

아래 ndarray를 slicing 하고 싶다.

array([[21, 22, 23],

       [33, 34, 35]])

#### cf> 하지만 행과 열을 동시에 slicing 불가능

```python
a[3:6][3:6]
```

## 3. Fancy Indexing

1> 내가 indexing한 부분들만 뽑아서 다시 이어주는 것 (discrete O) / 슬라이싱처럼 연속된 구간을 뽑아내는 것이 아니라 (continous X)

2> 대괄호 안에 또 대괄호를 넣는 방식으로 구현

### 1) Fancy Indexing 기본

```python
#a[2,5]
a[[2,5]]
```

### 2) Fancy Indexing 응용

```python
a[[2,5], [2,5]] # a[2,2]와 a[5,5]를 indexing 하는 것과 같다.
```

-----

아래 부분은 많이 어려운 내용입니다. 

여러분들에게 필수적으로 필요한 내용은 아니니 궁금하신 분들만 보시고 질문 받는 식으로 진행하겠습니다.

위의 fancy indexing과 유사합니다.

## 4. boolean Indexing (Masking)

### 1) boolean Indexing 1D

```python
boolean_1d = [True, True, False, False, True, True]
```

```python
a[boolean_1d]
```

### 2) boolean Indexing 2D

```python
boolean_2d = [True, True, False, False, True, True], [True, True, False, False, True, True]

a[boolean_2d]
```

```python
boolean_2d_tuple = ([True, True, False, False, True, True], 
                     [True, True, False, False, True, True])

a[boolean_2d_tuple]
```

```python
a = np.arange(6).reshape(2,3)
a[[True,False],[True,False,True]]
```
