# 4. 행과 열을 제거 및 결합

--------

## 1) 행과 열 제거 (drop)

### 0> drop에 shift+enter를 쳐보면 알겠지만 아래와 같은 parameter가 있습니다.

labels: 제거하고 싶은 대상

axis: 0(행을 제거), 1(열을 제거)
    
index: 선택한 행을 제거
    
columns: 선택한 열을 제거

#### 0-1> label + axis를 써서 제거할 수도 있고

#### 0-2> index 혹은 columns를 써서 제거할 수도 있습니다.

### 1> 행 제거

```python
df.head(3)
```

```python
df_row1 = df.drop(labels = 1) # axis=0이 default
df_row1
```

```python
df_row2 = df.drop(index=1) # axis=0이 default
df_row2
```

### 2> 열 제거

```python
df_column1 = df.drop(labels = 'Name', axis = 1)
df_column1
```

```python
df_column2 = df.drop(columns='Name')
df_column2
```

## 2) 행과 열 결합


열 concat, merge, join이 있지만 각각 복잡해 가장 간단한 것만 설명

행 append (주의점)

### 1> 열끼리의 결합 (join)

```python
df_col1 = df[['PassengerId', 'Survived']]
df_col1 
```

```python
df_col2 = df[['Fare', 'Ticket']]
df_col2
```

```python
df_new = df_col1.join(df_col2)
df_new
```

물론 이것 말고 더 많은 기능을 지원하는 'concat'과 'merge'가 있습니다.

원리는 유사한 부분이 있으니 이를 활용하실 분들은 이들을 공부해보시기 바랍니다.

### 2> 행끼리의 결합 (append)

```python
df_r1 = df.iloc[:99]
df_r1
```

```python
df_r2 = df.iloc[100:199]
df_r2
```

```python
df_r1.append(df_r2)
```
