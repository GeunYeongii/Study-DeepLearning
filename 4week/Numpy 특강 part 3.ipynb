## 1. axis란?

### 1) np.sum(axis = 0)

1> 행이 고정된 채 원소들끼리 더해지는 연산

2> 연산 이후 shape은 행 부분이 없어진다.

```python
a.shape
```

```python
a.ndim
```

```python
np.sum(a, axis=0)
```

```python
np.sum(a, axis=0).shape
```

```python
np.sum(a, axis=0).ndim
```

### 2) np.sum(axis = 1)

1> 열이 고정된 채 원소들끼리 더해지는 연산

2> 연산 이후 shape은 열 부분이 없어진다.

```python
np.sum(a, axis=1)
```

```python
a.shape # (4,5) -> (4,)
```

```python
np.sum(a, axis=1).shape
```

```python
np.sum(a, axis=0).shape
```

```python
np.sum(a, axis=1).ndim
```

아쉬운 점: sum을 하면 차원이 줄어드는 현상이 벌어진다.
    
np.sum(a, axis=1) => array([10, 35, 60, 85])말고 

np.sum(a, axis=1) => array([[10],

                          [35],
                           
                          [60],
                           
                          [85]])  이렇게 나오면 좋을 것 같을 때 어떡할까?

## 2.keepdims

말 그대로 차원(dimension)이 유지(keep)되는 연산이다.

```python
np.sum(a, axis=0, keepdims = False)
#np.sum(a, axis=0)
```

```python
np.sum(a, axis=0, keepdims = True)
```

더 신기한 건 이 부분이다.

```python
np.sum(a, axis=1, keepdims = False)
```

```python
np.sum(a, axis=1, keepdims = True)
```

```python
np.sum(a, axis=1, keepdims = True).ndim
```

```python
np.sum(a, axis=1, keepdims = False).ndim
```

keepdims는 앞으로 딥러닝하면서 많이 나오는 내용은 아니지만

axis 개념은 빈출되는 내용이니 여러번 봐주시기 바랍니다.
