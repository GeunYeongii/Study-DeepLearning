# 1) 행

- ndarray처럼 df[행, 열] 방식으로 indexing과 slicing이 불가능합니다.

- <font color="red"> loc(location의 약자) 혹은 iloc(integer location의 약자)를 사용해야합니다.

## 1-1) loc

#### index가 정수형(integer)가 아닌 경우에 사용

- df.loc를 쓰게 되면 ndarray indexing & slicing 사용하듯 쓸 수 있습니다.

- 다른 점은 <font color="red">'[start:end]= start이상 end이하'</font>가 된다는 점입니다.

## 1-2) iloc

#### index가 정수형인 경우에 사용한다.

- 나머지 사용법은 loc와 유사합니다.

```python
dic4 = {'age': [20,24,27], 'orgin': ['Germany', 'China', 'Korea'], 'major':['computer science', 'math', 'economics']}
df4 = pd.DataFrame(data = dic4, index = ['Harry', 'Linda', 'Kim'] )
df4
```

### 1> Indexing

> #### 1-1> 1개 Indexing (Series로 출력)

```python
df4.loc['Harry']
```

#### cf> 만약 Harry의 data와 Linda의 data를 모두 가져오려면?

이건 안 됩니다.

```python
# 이건 numpy에서도 error

df4.loc['Harry', 'Linda']
```

> #### 1-2> 1개 Indexing (DataFrame으로 출력) [fancy indexing]

```python
df4.loc[['Harry']]
```

> #### 1-3> n개 Indexing (DataFrame으로 출력) [fancy indexing]

```python
df4.loc[['Harry', 'Linda']]
```

불편하게 fancy indexing을 쓰지 말고 slicing을 배워보겠습니다.

### 2> Slicing

- df.loc를 쓰게 되면 ndarray indexing & slicing 사용하듯 쓸 수 있습니다.

- 다른 점은 <font color="red">'[start:end]= start이상 end이하'</font>가 된다는 점입니다.

```python
df4.loc['Harry':'Kim']
```

```python
df4.loc['Linda':'Kim', 'age':'orgin']
```

### cf> 예외사항

```python
df4[0:1]
```

-----

## 2) 열

- DataFrame에서 하나의 column(열)은 series입니다.

- <font color="red"> column의 name으로 indexing & slicing 한다.

```python
URL = './heart.csv'
df = pd.read_csv(URL)
df.head() # 위의 5개 행만 slicing
```

### 1> Indexing
> #### 1-1> 1개 Indexing (Series로 출력)

```python
# 1번

df['chol']
```

```python
# 2번
 
df.chol
```

> #### 1-2> 1개 Indexing (DataFrame으로 출력) [fancy indexing]

```python
df[['chol']]
```

> #### 1-3> n개 Indexing (DataFrame으로 출력) [fancy indexing]

```python
df[['chol', 'age']]
```

### 2> Slicing

```python
df['cp':'ca'] # column이 행 부분을 의미하는 것이 아니므로 아래처럼 slicing을 해야합니다.
```

```python
df.loc[:, 'cp':'ca']
```
