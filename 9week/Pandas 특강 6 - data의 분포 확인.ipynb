# 6. data의 분포 확인

아마 9주차 수업을 들으셨다면 class의 분포나 feature의 분포를 확인하는 것이 중요하다고 느끼셨을 것입니다.

이들을 pandas로 쉽게 구현해보겠습니다.

--------

## 1) data의 통계 정보 확인 (describe)

describe는 각 columns에 대한 통계정보를 볼 수 있습니다.

```python
df.describe()
```

하지만 이것만으로는 여러분들이 분포를 완벽하게 파악하기 어렵습니다.

## 2) 각 column이 어떤 값들을 갖는지 확인하고 싶다. (unique)

```python
df['Survived'].unique()
```

## 3) 각 column이 갖는 값들의 개수를 알고 싶다. (value_counts)

저희는 class가 balance를 가지는지 알고 싶습니다. 이 때 쓰는 것이 value_counts입니다.

```python
df['Survived'].value_counts()
```

약간 imbalance한 상황입니다.
