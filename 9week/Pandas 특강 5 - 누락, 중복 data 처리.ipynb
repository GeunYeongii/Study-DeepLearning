# 5. 누락, 중복 data 처리

--------

## 1) 누락 data 확인

### 1> .info()

- DataFrame에 대한 요약 정보가 담겨 있으며

- 한 번에 null의 개수를 확인할 수 있습니다.

```python
df.info()
```

Age, Cabin, Embarked에 null값이 존재하는 걸 확인할 수 있습니다.

### 2> .isna(), .notna()

#### 누락 data를 찾는 직접적인 방법이며, 전체 DataFrame을 출력합니다. (요약적이지는 않습니다.)

- isna(): 누락 data에서 True를 반환 (== isnull() )

- notna(): 누락 data에서 False를 반환 (== notnull() )

```python
df.isna()
```

------

## 2) 누락 data 채워 넣기 (fillna)

### 1> 0으로 채워 넣기

```python
df_fill_0 = df.fillna(0)
df_fill_0.info()
```

### 2> 다르게 채워 넣기

```python
import numpy as np
```

```python
each = {'Age':np.mean(df.Age), 'Embarked':0 }
df_fill_each = df.fillna(value = each)
df_fill_each.info()
```

### 3> 누락 column 제거

```python
df.drop('Cabin', axis=1)
```

------

## 3) 중복 data 확인 (duplicated)

```python
df.duplicated()
```

```python
df[df.duplicated()]
```

-------

## 4) 중복 data 제거 (drop_duplicates)

```python
df_remove_dup = df.drop_duplicates()
df_remove_dup
```

```python
df_remove_dup[df_remove_dup.duplicated()]
```
